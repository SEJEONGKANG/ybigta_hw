```python
pd.merge(df_x, df_y, how='left', on='quality')
```

- 데이터 : deliver.csv (날씨정보에 따른 배달품목별 주문량 데이터) / '내 드라이브'에 사본 저장할 것.

1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라.
(계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)

2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라.
(전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)

```python
data2 = pd.read_csv('/content/drive/MyDrive/연세/YBIGTA/220714_numpypandas/deliver.csv')
data2.head()
```

```python
#1. '날짜' 컬럼을 보고, '월' 컬럼과 '계절' 컬럼을 새로 추가하라. (계절 컬럼 : 3-5월 : 봄, 6-8월 : 여름, 9-11월: 가을, 12-2월 : 겨울)
def add_season(mon):
    if int(mon[5:7]) in (3,4,5):
        return '봄'
    elif int(mon[5:7]) in (6,7,8):
        return '여름'
    elif int(mon[5:7]) in (9,10,11):
        return '가을'
    else:
        return '겨울'

data2['월']= data2['날짜'].apply(lambda x: x[5:7]+"월")
data2['계절']=data2['날짜'].apply(add_season)
data2.head(30)
```

```python
#2. '미세먼지(㎍/㎥)' 컬럼에서 null값의 개수를 확인하고, null값을 전체 평균으로 대체하라. (전체 평균은 null값이 아닌 값들의 평균으로 계산된다.)
print(data2['미세먼지(㎍/㎥)'].isnull().sum())
```

```python
data2.loc[data2['미세먼지(㎍/㎥)'] != data2['미세먼지(㎍/㎥)'], '미세먼지(㎍/㎥)'] = data2['미세먼지(㎍/㎥)'].mean()
data2.head(30)
```
