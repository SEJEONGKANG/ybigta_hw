# 1. Imputation(결측값 대체)


```python
titanic.isnull().sum() #결측치 확인
```

```python
titanic2 = titanic.drop(['Cabin'], axis = 1) # cabin은 418 행 중 327 행이 결측치이므로 삭제
titanic3 = titanic2.fillna(titanic2.median()) # age와 fare 은 중앙값으로 결측값 대체
```

```python
titanic3.isnull().sum()
```

```python
titanic3
```

# 2. 지표변수 (Indicator variables)


```python
titanic['Age'].describe()
```

```python
titanic4 = titanic
age = round(titanic['Age'])  # min 값이 0.17이므로 age 열의 값을 정수로 바꿈
titanic4['Age'] = age
```

```python
titanic4['Age'].describe()
```

```python
oldness = []
for a in titanic4['Age']:
    if (a <= 25):
        oldness.append('young')
    elif (a <= 37):
        oldness.append('middle')
    else:
        oldness.append('old')

titanic4['oldness'] = oldness
```

```python
titanic4
```

# 3. Feature Split


```python
titanic5 = titanic
First_Name = []

for d in titanic['Name'].str.split(","):
    First_Name.append(d[0])
titanic5['First_Name'] = First_Name
```

```python
titanic5
```

# 4. 스케일링 Scaling

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
mm_scaler = MinMaxScaler()
sd_scaler = StandardScaler()
```

```python
titanic6 = titanic

# MinMax Scaling
mm_scaler = mm_scaler.fit_transform(titanic6['Fare'].values.reshape(-1,1))
titanic6['mm_scaled_fare'] = mm_scaler

# Standard Scaling
sd_scaler = sd_scaler.fit_transform(titanic6['Fare'].values.reshape(-1,1))
titanic6['sd_scaled_fare'] = sd_scaler
```

```python
titanic6
```
