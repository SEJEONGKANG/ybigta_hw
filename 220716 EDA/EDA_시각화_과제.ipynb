### 2) 결측치 확인

- #### <font color='orange'> Q2. 각 컬럼의 결측치를 확인하고, 결측치가 많은 순서대로 결측치 비율과 함께 표로 나타내보세요.
<font color='orange'> (결측치 0개인 컬럼은 나타내지 마세요)

```python
# Q2
df.info()
```

```python
nullratio = df.isnull().sum() / len(df) * 100
nullratio2 = nullratio[nullratio > 0]
df2 = pd.DataFrame(nullratio2, columns = ['null_ratio'])
df2 = df2.sort_values(by='null_ratio', ascending = False)
df2
```

- #### <font color='orange'> Q3. 결측치가 있는 컬럼과 결측치 비율을 보기쉽게 그래프로 나타내 보세요.

```python
# Q3
plt.figure(figsize = (30,10))
ax = sns.barplot(x = df2.index, y = df2.iloc[:,0])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 45)
plt.show()
```

- #### <font color='orange'> Q4. 결측치가 적은 beds 컬럼을 좀 더 자세히 살펴보겠습니다.

- #### <font color='orange'> Q6. 위 그래프의 세로축이 갯수가 아닌 비율을 나타내도록 재구성해보세요.

- #### <font color='orange'> Q7. 각 지역별로 room_type의 비율이 어땠는 지 알고 싶습니다.

- #### <font color='orange'> Q8. 각 지역별로 가격의 분포가 어떤지 알고 싶습니다.
